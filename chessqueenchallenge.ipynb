```
import random
import pandas as pd

class ChessQueen(object):

  def __init__(self,totalPopulation,crossOver,mutation):
    self.mutation=mutation
    self.totalPopulation=totalPopulation
    self.crossOver=crossOver


  def initialPopulation(self,target,totalPopulation,number_list):
    population_data_list=[]
    fitnessScore_list=[]
    for population_iteration in range(totalPopulation):
      fitnessScore=0
      random_generated_data=[]
      for target_position in range(len(target)):
        random_generated_value=secure_random.choice(number_list)
        if (str(random_generated_value) == target[target_position]):
          fitnessScore=fitnessScore+1
        random_generated_data.append(random_generated_value)
      population_data_list.append(random_generated_data)
      fitnessScore_list.append(fitnessScore)
    return population_data_list, fitnessScore_list

  def generateProbability(self,target,totalPopulation,populationData,fitnessData):
    probabilityDist=[]
    for outloop in range(totalPopulation):
      probabilityDist.append(fitnessData[outloop]/len(target))
    probDataFrame=pd.DataFrame({'String':populationData,'FitnessScore':fitnessData,'probability':probabilityDist})
    probDataFrame=probDataFrame.sort_values(['probability'],ascending=False)
    probDataFrame=probDataFrame.reset_index(drop=True)
    return probDataFrame

  def getFitnessScore(self,input,target):
    fitnessScore=0
    for inloop in range(len(target)):
      if (str(input[inloop]) == target[inloop]):
        fitnessScore = fitnessScore + 1
    return fitnessScore
  
  def fitnessCheck(self,fitnessScoreValue,target):
    if (fitnessScoreValue==len(target)):
      return True
    else:
      return False

  def crossOverAndMutation(self,crossOverPoint,number_list,target,populationData,first_list,second_list):
    child1 = first_list[0:crossOverPoint]+second_list[crossOverPoint:]
    child2 = second_list[0:crossOverPoint]+first_list[crossOverPoint:]    
    child1[random.randint(0,len(target)-1)] = secure_random.choice(number_list)
    child2[random.randint(0,len(target)-1)] = secure_random.choice(number_list)
    populationData.append(child1)
    populationData.append(child2)
    return populationData

  def viewString(self,input):
    return (''.join(elem for elem in [str(each) for each in input]))


  def main(self):
    target="6 4 2 0 5 7 1 3"
    number_list=[x for x in range(0,10,1)]
    number_list.append(' ')
    populationData, fitnessData=self.initialPopulation \
      (target,self.totalPopulation,number_list)
    probDataFrame=self.generateProbability(target,self.totalPopulation,populationData,fitnessData)
    crossOverPoint = int(self.crossOver*len(target))
    generationCount = 1000
    for loop in range(generationCount):
      draw=[]
      draw.append(probDataFrame[0:1]["String"].values[0])
      draw.append(probDataFrame[1:2]["String"].values[0])
      if ((self.getFitnessScore(draw[0],target)) == len(target) | (self.getFitnessScore(draw[1],target)) == len(target)):
        print (self.viewString(draw[0]) ,  " and " , self.viewString(draw[1]))
        for itr in [0,1]:
          result=self.fitnessCheck((self.getFitnessScore(draw[itr],target)),target)
          if (result):
            print (" qconfig genetic sequence : ", self.viewString(draw[itr]))
            return self.viewString(draw[itr])
      populationData=self.crossOverAndMutation(crossOverPoint,number_list,target,populationData,draw[0],draw[1])
      self.totalPopulation=len(populationData)
      fitnessData=[]
      for outloop in range(self.totalPopulation):
        fitnessScore=self.getFitnessScore(populationData[outloop],target)
        fitnessData.append(fitnessScore)
      probDataFrame=self.generateProbability(target,self.totalPopulation,populationData,fitnessData)
      
      print("Generation : ", loop , " Parent1 : ", self.viewString(draw[0]) , \
             " fitscore Parent1 : " , self.getFitnessScore(draw[0],target) ,
            " and Parent2 : " , self.viewString(draw[1]), \
            " fitscore Parent2 : " , self.getFitnessScore(draw[1],target))
    





secure_random = random.SystemRandom()
cq = ChessQueen(totalPopulation=150,crossOver=0.5,mutation=0.01)
cq.main()
```
